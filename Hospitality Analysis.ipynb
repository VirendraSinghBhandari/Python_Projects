# Data Transformation

```python
df_agg_booking
```

```python
df_agg_booking['Occ_Pct'] = df_agg_booking['successful_bookings']/df_agg_booking['capacity']
df_agg_booking.head()
```

```python
df_agg_booking['Occ_Pct'] = df_agg_booking['Occ_Pct'].apply(lambda x: round(x*100, 2))
df_agg_booking.head()
```

# 1. What is an average occupancy rate in each of the room category?

```python
df_agg_booking.groupby('room_category')['Occ_Pct'].mean().round(2)
```

```python
df_rooms
```

```python
df = pd.merge(df_agg_booking, df_rooms, left_on = 'room_category', right_on = "room_id")
df.head()
```

```python
df.groupby('room_class')['Occ_Pct'].mean().round(2)
```

```python
df.groupby('room_class')['Occ_Pct'].mean().round(2).plot()
```

```python
df.drop('room_id', axis = 1, inplace = True)
df.head()
```

# 2. Print average occupancy rate per city?

```python
df_hotels
```

```python
df= pd.merge(df,df_hotels, on = 'property_id')
df.head()
```

```python
df.groupby('city')['Occ_Pct'].mean().plot(kind="bar")
```

# 3. When was the occupancy better? Weekdays or Weekend?

```python
df_date
```

```python
df= pd.merge(df,df_date, left_on= 'check_in_date', right_on = 'date')
df.head()
```

```python
df.groupby('day_type')['Occ_Pct'].mean().plot(kind = 'barh')
```
